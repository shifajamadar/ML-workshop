```python
from numpy.random.mtrand import f
from matplotlib.lines import lineStyles
#How has startup funding treand changed over the years?
#data visualization using matplotLib Library

import matplotlib.pyplot as plt

demo=data.groupby('year')['amount'].sum()
demo
plt.plot(demo.index,demo.values, marker="o",markerfacecolor="red",color="purple",linestyle="--",linewidth=1.5,label="Total Amount")
plt.xlabel("Year",color="red",fontsize=12)
plt.ylabel("Amount",color="blue",fontsize=12)
plt.grid(color="skyblue",linestyle="--",linewidth=1.1)
plt.title("Startup Funding Trend",color="red",fontsize=16)
plt.legend()
plt.show()
```

```python
# which city have the highest number of startups?
# top 10 cities using head function
import matplotlib.pyplot as plt

demo=data['city'].value_counts().head(10)
color=["#8140DC","Purple","blue","#FF3483","red","orange","#B34A44","#E77B49","#E22F80","#FFB6A6"]
plt.bar(demo.index,demo.values,color=color,alpha=0.4)
plt.xlabel("City",color="blue",fontsize=10)
plt.ylabel("Number of Startups",color="blue",fontsize=10)
plt.grid(color="skyblue",linestyle="--",linewidth=1.2)
plt.title("Top 10 Cities with Highest Number of Startups",color="red",fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
## which startup sectors have the highest number?
import matplotlib.pyplot as plt

demo=data['vertical'].value_counts().head(10)
color=["#8140DC","Purple","blue","#FF3483","red","orange","#B34A44","#E77B49","#E22F80","#FFB6A6"]
plt.barh(demo.index[::-1],demo.values[::-1],color=color,alpha=0.4,label="Number of Startups",edgecolor="black")
plt.xlabel("Number of Startups",color="blue",fontsize=10)
plt.ylabel("Startup Top Sector",color="blue",fontsize=10)
plt.xlabel("Number of Startups",color="blue",fontsize=10)
plt.grid(axis="x",color="gray",linestyle="--",linewidth=1.3)
plt.title("Top 10 Startup Sectors with Highest Number of Startups",color="red",fontsize=11)
plt.figure(figsize=(10,10))
plt.show()
```

```python
#what is the distribution of different funding round?
import matplotlib.pyplot as plt

demo=data['round'].value_counts().head(7)
color=["#8140DC","Purple","blue","#FF3483","red","orange","#B34A44"]
plt.pie(demo.values,labels=demo.index,autopct="%1.1f%%",colors=color)
plt.title("round Distrubtion")
plt.figure(figsize=(10,10))
plt.show()
```

```python
#How are startup funding amount distributed ?  Draw histrogram chat
import matplotlib.pyplot as plt
import pandas as pd

demo=data['amount'].value_counts().head(10)
plt.hist(demo,bins=20,color="red",edgecolor="black")

```

```python
#the relationship between startup funding amount and year? using scatter.
import matplotlib.pyplot as plt

plt.scatter(data['year'],data['amount'],color="red")
plt.title("Relationship between startup funding amount and year")
plt.xlabel("year")
plt.ylabel("amount")
plt.grid(color="skyblue",linestyle="--")
plt.show()
```

```python
#Which funding round occurs most frequently? using seaborn
import seaborn as sns

demo=data['round'].value_counts().head(8).index
sns.countplot(data,y='round',order=demo,palette='viridis')
plt.title("Most frequently Funding Round")
plt.xlabel("Count")
plt.ylabel("Funding Round")
plt.figure(figsize=(10,6))
plt.show()
```
